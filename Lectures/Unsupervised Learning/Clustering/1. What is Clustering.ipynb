```python
## For data handling
import numpy as np

## For plotting
import matplotlib.pyplot as plt
from seaborn import set_style

## This sets the plot style
## to have a grid on a dark background
set_style("white")
```

Another common unsupervised learning task in <i>clustering</i>. In clustering we look to identify groupings of similar data points in otherwise unlabeled data, $X$. This could be for any number of reasons including:
- Identifying market segments,
- Locating similar product users,
- etc.

Consider the following data, `X`.

```python
X = np.zeros((200,2))

X[:100,:] = np.random.randn(100,2)
X[100:,:] = np.array([4,4]) + np.random.randn(100,2) 
```

```python
plt.figure(figsize=(8,8))

plt.scatter(X[:,0], X[:,1])

plt.xlabel("$X_1$", fontsize=16)
plt.ylabel("$X_2$", fontsize=16)

plt.show()
```

A clustering algorithm may look at this data and produce something like the following image.

```python
plt.figure(figsize=(8,8))

plt.scatter(X[:100,0], X[:100,1], c='r', s=40, label="Cluster 1")
plt.scatter(X[100:,0], X[100:,1], c='b', s=100, marker='+', label="Cluster 2")


plt.xlabel("$X_1$", fontsize=16)
plt.ylabel("$X_2$", fontsize=16)

plt.legend(fontsize=14)

plt.show()
```
